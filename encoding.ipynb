```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data={'Color':['Red','Blue','Green','Blue','Red']}
df4=pd.DataFrame(data)
print("Original data")
print(df4)
#Apply label encoding
le=LabelEncoder()
df4['Color_encoded']=le.fit_transform(df4['Color'])
print("After label encoding")
print(df4)
```

```python
#Apply one hot encoding
df4_encoded=pd.get_dummies(df4,columns=['Color'],dtype=int)
print("After One-Hot Encoding")
print(df4_encoded)
```
