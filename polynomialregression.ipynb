Polynomial Regression is a type of Linear Regression where the relationship between the independent variable X and dependent variable y is non-linear.
It fits a curve instead of a straight line.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv('polynomial.csv')
print(df.head())
```

```python
X=df[['Temperature']]
y=df['Pressure']
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
# X_train,y_train,X_test,y_test=train_test_split(X,y,test_size=0.2)
# X_train_poly=poly.fit_transform(X_train)
# X_test_poly=poly.transform(X_test)
```

```python
X_poly=poly.fit_transform(X)
model=LinearRegression()
model.fit(X_poly,y)
```

```python
y_pred=model.predict(X_poly)
```

```python

plt.scatter(X,y,color='blue',label='Data Points')
plt.plot(X,y_pred,color='red',label='Polynomial Regression Line')
plt.xlabel('Temperature')
plt.ylabel('Pressure')
plt.legend()
plt.show()
```
