```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("C:\\Users\\hp\\Documents\\Python\\wine.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('OD',axis=1), df['OD'], test_size=0.2, random_state=42)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled
X_test_scaled
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)
X_train_scaled_df.head()
X_test_scaled_df.head()
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.histplot(X_train['Alcohol'], kde=True, color='blue', bins=30)
plt.title('Before Normalization - Alcohol')
plt.subplot(1,2,2)
sns.histplot(X_train_scaled_df['Alcohol'], kde=True, color='green', bins=30)
plt.title('After Normalization - Alcohol')
plt.show()
```
