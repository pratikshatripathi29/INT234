```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 4))
plt.bar(missingvalues.index, missingvalues.values)
plt.title('Missing Values per Column')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#replacing missing values with mean for numrecial column and with mode for categorical column
df['age']=df['age'].fillna(value=df['age'].mean())
print(df['age'])
df['total_spent']=df['total_spent'].fillna(value=df['total_spent'].mean())
print(df['total_spent'])
df['country']=df['country'].fillna(value=df['country'].mode())
print(df['country'])
```

```python
q1=df['total_spent'].quantile(0.25)
q3=df['total_spent'].quantile(0.75)
iqr=q3-q1;
print("iqr: ",iqr)
lb=q1-1.5*iqr;
ub=q3+1.5*iqr;
print("Lower bound: ",lb)
print("Upper bount: ",ub)
```

```python
#Outliers>>>>>>>>>>>>>>>>>>>>>>>>
plt.boxplot(df["total_spent"])
plt.title("Boxplot: total_spent")
plt.ylabel("total_spent")
plt.tight_layout()
plt.show()
```

```python
#Applying Min-Max Scaling and Standarization
from sklearn.preprocessing import MinMaxScaler,StandardScaler
num_cols=df.select_dtypes(include=['number']).columns
print(num_cols)
print("Numeric columns:",num_cols.tolist())
```

```python
#Creating scaler object
normalizer=MinMaxScaler()
#Apply normalization
normalized=df.copy()
normalized[num_cols]=normalizer.fit_transform(df[num_cols])
print(normalized.head())
```

```python
#create a standard sacler object
scaler=StandardScaler()
#Apply standard scaling
standardized=df.copy()
standardized[num_cols]=scaler.fit_transform(df[num_cols])
print(standardized.head())
print("Before Scaling")
print(df[num_cols].describe())
print(df[num_cols])
print("After Normalization\n")
print(normalized[num_cols].describe())
print("After Standarization\n")
print(standardized[num_cols].describe())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['country']=le.fit_transform(df['country'])
print("After label encoding")
print(df)
```
