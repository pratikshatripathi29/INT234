```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
df=pd.read_csv("C:\\Users\\hp\\Documents\\Python\\wine.csv")
X=df.drop("Wine" ,axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, df["Wine"], test_size=0.2)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
linear_model= SVC(kernel='linear')
linear_model.fit(X_train,y_train)
linear_pred=linear_model.predict(X_test)
print("Linear SVM")
print(confusion_matrix(y_test,linear_pred))
print(classification_report(y_test,linear_pred))
print("Accuracy:",accuracy_score(y_test,linear_pred))
```

```python
rbf_model= SVC(kernel='rbf')
rbf_model.fit(X_train,y_train)
rbf_pred=rbf_model.predict(X_test)
print("RBF SVM")
print(confusion_matrix(y_test,rbf_pred))
print(classification_report(y_test,rbf_pred))
print("Accuracy:",accuracy_score(y_test,rbf_pred))
```

```python
poly_model= SVC(kernel='poly')
poly_model.fit(X_train,y_train)
poly_pred=poly_model.predict(X_test)
print("Polynomial SVM")
print(confusion_matrix(y_test,poly_pred))
print(classification_report(y_test,poly_pred))
print("Accuracy:",accuracy_score(y_test,poly_pred))
```

```python
sigmoid_model= SVC(kernel='sigmoid')
sigmoid_model.fit(X_train,y_train)
sigmoid_pred=sigmoid_model.predict(X_test)
print("Sigmoid SVM")
print(confusion_matrix(y_test,sigmoid_pred))
print(classification_report(y_test,sigmoid_pred))
print("Accuracy:",accuracy_score(y_test,sigmoid_pred))
```
