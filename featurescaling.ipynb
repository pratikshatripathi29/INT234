```python
import pandas as pd
df=pd.read_csv('Iris.csv')
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

Xdf = df.drop(['Species'], axis=1)
scalar=StandardScaler()
X_scaled=scalar.fit_transform(Xdf)
print(X_scaled)

```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_minmax_scaled = scaler.fit_transform(Xdf)
print(df_minmax_scaled)
```
